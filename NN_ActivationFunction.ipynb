```python
import matplotlib.pyplot as plt
import numpy as np
```

LOGISTIC OR SIGMOID

```python
# np.linspace(-10,10,20)
x=np.arange(-10,10,1)
z = 1/(1+np.exp(-x))

plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("Logistic(x)")

plt.show()
```

HYPERBOLIC TAGENT

```python
x=np.arange(-10,10,1)
z = (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("HyperbolicTangent(x)")

plt.show()
```

LINEAR

```python
x=np.arange(-10,10,1)
z = x

plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("Linear(x)")

plt.show()
```

SIGN(SIGNUM)

```python
x=np.arange(-10,10,1)
z = []

for i in x:
  if i < 0:
    z.append(-1)
  elif i == 0:
    z.append(0)
  else:
    z.append(1)

print(x)
print(z)

plt.plot(z)
plt.xlabel("x")
plt.ylabel("Signum(x)")

plt.show()
```

UNIT STEP (HEAVISIDE)

```python
x=np.arange(-10,10,1)
z = []

for i in x:
  if i < 0:
    z.append(0)
  elif i == 0:
    z.append(0.5)
  else:
    z.append(1)

print(x)
print(z)

plt.grid()
plt.xlim((-1,20))
plt.ylim((-1.5,2))

plt.plot(z)
plt.xlabel("x")
plt.ylabel("Heaviside(x)")

plt.show()
```

PIECE-WISE LINEAR

```python
x=np.arange(-10,10,1)
z = []

for i in x:
  if i >= 1/2:
    z.append(1)
  elif i < 1/2 or i > -1/2:
    z.append(i+(1/2))
  else: # i <= - 1/2
    z.append(0)

print(x)
print(z)

plt.grid()
plt.xlim((-11,11))
plt.ylim((-11,11))
plt.plot(z)
plt.xlabel("x")
plt.ylabel("PieceWiseLinear(x)")

plt.show()
```
